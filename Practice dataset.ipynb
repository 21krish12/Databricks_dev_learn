```python
df=df_tx.join(df_cat,df_tx.category_id==df_cat.cat_id,'inner').select('id','name','category_id','category_name','amount','status','department')
```

```python
from pyspark.sql.functions import col,sum,max,min,count
df.groupBy("category_id").agg(sum(col('amount')).alias('total_amount'),max(col('amount')).alias('max_amount'),min(col('amount')).alias('min_amount'),count(col('id')).alias('No_of_transactions')).display()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col
from pyspark.sql.functions import row_number,dense_rank,rank,lag,lead
w = Window.partitionBy("name").orderBy(col("amount").desc(),col('category_id').asc())

(df_tx.withColumn('rn',row_number().over(w)) \
    .withColumn('rnk',rank().over(w)) \
    .withColumn('drnk',dense_rank().over(w)) \
    .withColumn('lag1',lag('name',1).over(w)) \
    .withColumn('lead1',lead('name',1).over(w)) \
    .orderBy(col('name').asc()) \
    .display())
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Window: per name, highest amount first, then lowest category_id on ties
w = Window.partitionBy("name").orderBy(F.col("amount").desc(), F.col("category_id").asc())

# Fixed & improved version
(df_tx.withColumn("rn",   F.row_number().over(w)).withColumn("rnk",  F.rank().over(w)).withColumn("drnk", F.dense_rank().over(w)).withColumn("lag_name",  F.lag("name", 1).over(w)).withColumn("lead_name", F.lead("name", 1).over(w)) \
     .orderBy(F.col("name").asc()) \
     .show(truncate=False)  )        # or .display() in Databricks
```
