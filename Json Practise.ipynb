```python
from pyspark.sql.functions import from_json, explode
from pyspark.sql.types import *
df=spark.read.option('multiline',True).json('/Volumes/bronze/stage/raw/Products_netsuite.json')

schema = ArrayType(
    StructType([
        StructField("links", ArrayType(StringType()), True),
        StructField("averagecost", StringType(), True),
        StructField("createddate", StringType(), True),
        StructField("description", StringType(), True),
        StructField("isinactive", StringType(), True),
        StructField("lastmodifieddate", StringType(), True),
        StructField("netsuiteinternalid", StringType(), True),
        StructField("prodcut_code", StringType(), True),
        StructField("productclass", StringType(), True),
        StructField("productdisplaynamecode", StringType(), True),
        StructField("productnamenumber", StringType(), True),
        StructField("unitprice", StringType(), True),
        StructField("updatedsince", StringType(), True)
    ])
)

df1 = df.withColumn(
    "response_parsed",
    from_json("response", schema)
)

final_df = df1.select(explode("response_parsed").alias("item")).select("item.*")

final_df.display()
```
